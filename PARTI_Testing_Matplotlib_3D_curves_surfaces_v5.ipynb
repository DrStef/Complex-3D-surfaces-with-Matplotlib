## Helicoid - Catenoid

Equations can be found in Wikipedia:  https://en.wikipedia.org/wiki/Catenoid  

from Wiki: A parametrization of such a deformation (Helicoid transformation) is given by the system

\begin{aligned}x(u,v)&=\cos \theta \,\sinh v\,\sin u+\sin \theta \,\cosh v\,\cos u\\y(u,v)&=-\cos \theta \,\sinh v\,\cos u+\sin \theta \,\cosh v\,\sin u\\z(u,v)&=u\cos \theta +v\sin \theta \end{aligned}

for  $(u,v) \in [-\pi ,\pi] \times R$

and where:

If $\theta= \pi/2$, the surface is a catenoid.  <br>
If $\theta = 0 $  the surface is a left-handed helicoid <br>
If $\theta = \pi$ the surface is a right-handed helicoid. <br>
At some point will build an animation from helicoid to catenoid.  

```python
theta=0     # Helicoid
theta=np.pi//4       # HElICOID
theta=np.pi/2   # CATENOID
resol=200

#------------   [-pi, pi[   
#limt= np.pi
#uu=np.arange(-limt, limt, (2*limt+2)/resol, dtype=float)
limt= np.pi
uu=np.arange(0, 2*limt+limt/resol, 2*limt/resol, dtype=float)
#-----------    v in R   -inf, +inf
infini= 1.5
vv=np.arange(-infini, infini, 2*infini/resol , dtype=float)
u,v = np.meshgrid(uu, vv)

xc = np.cos(theta)*np.sinh(v)*np.sin(u) + np.sin(theta)*np.cosh(v)*np.cos(u)
yc = -np.cos(theta)*np.sinh(v)*np.cos(u) + np.sin(theta)*np.cosh(v)*np.sin(u)
zc = u*np.cos(theta) + v*np.sin(theta)

#----------------------------------
#xc= c * np.cosh(t/c)*np.cos(theta)
#yc= c * np.cosh(t/c)*np.cos(theta)
#zc= np.sinh(t/c)

fig = plt.figure(figsize = (10,10))
ax = plt.axes(projection='3d')
ax.grid()
#ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='jet', edgecolor='none', theta=0.4)
ax.plot_surface(xc, yc, zc, cmap='jet', edgecolor='none', alpha=1)
#ax.plot3D(xc,yc,zc, color=(( 0.2, 0.2, 1, 0.8)),linewidth=2)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

```

## Toroidal surfaces 


<span style="color:#4169E1"> Equations can be found on Paul Bourque's website.     http://paulbourke.net/geometry/toroidal/

<span style="color:#4169E1"> The torus parametric equations are a classic. 

<span style="color:#4169E1"> Paul Bourque's remarks regarding super-toroids (in italic, green): <br>
<i> <span style="color:#446600"> "The super-toroid is a family of geometric primitives based on the torus. Let: <br>
* <i> <span style="color:#446600"> r0 the radius of the inner ring 
* <i>  <span style="color:#446600"> r1 the radius of the outer ring <br>
    
<i><span style="color:#446600"> as illustrated below where theta and phi range from 0 to 2π.
<i> <span style="color:#446600"> The equation for the super-toroid is the same as that for the torus except that the sin and cosine terms are raised to powers.</i>
<span style="color:#446600">   <i> It is different values of these powers which give rise to a family of 3D shapes all basically toroidal in shape. The value of $n_{1}$ determines the shape of the torus ring, $n_{2}$ determines the shape of the cross section of the ring. Note that with the normal definition of power the above equations are only defined for a single quadrant, namely that where the cosine and sine terms are positive. When creating the super-torus in practice one has two choices, either replicate the result in the one quadrant to the other three with the correct mirror operations, or define: </i>
    
<span style="color:#446600"> $$x^{n} = sign(x)abs(x)^{n}$$    
    
<span style="color:#4169E1"> Note that, depending on values of $n_{1}$ and $n_{2}$,  this tranform is continuous **but not necessarily differentiable**. It ensures the continuity of the toroidal surface at the interfaces of each quadrant, but discontinuity of tangents can occur: sharp corners,... 
  
<span style="color:#4169E1"> Examples of the super-toroids generated for different values of $n_{1}$ and $n_{2}$ are shown below:   

```python
# Torus Parameters 
R=1
r=0.2
#a=0.4
#---- 
resol=200
u =np.arange(0, 4*np.pi, np.pi/resol)
v =np.arange(0, 4*np.pi, np.pi/resol)

uu,vv=np.meshgrid(u,v)


X=(R+r*np.cos(vv))*np.cos(uu)
Y=(R+r*np.cos(vv))*np.sin(uu)
Z=r*np.sin(vv)

n1=3
n2=3
X1=(R+r*np.sign(np.cos(vv))*np.power(abs(np.cos(vv)),n2))*np.sign(np.cos(uu))*np.power(abs(np.cos(uu)),n1)
Y1=(R+r*np.sign(np.cos(vv))*np.power(abs(np.cos(vv)),n2))*np.sign(np.sin(uu))*np.power(abs(np.sin(uu)),n1)
Z1=r*np.sign(np.sin(vv))*np.power(abs(np.sin(vv)),n2)

n1=3
n2=0.5
X2=(R+r*np.sign(np.cos(vv))*np.power(abs(np.cos(vv)),n2))*np.sign(np.cos(uu))*np.power(abs(np.cos(uu)),n1)
Y2=(R+r*np.sign(np.cos(vv))*np.power(abs(np.cos(vv)),n2))*np.sign(np.sin(uu))*np.power(abs(np.sin(uu)),n1)
Z2=r*np.sign(np.sin(vv))*np.power(abs(np.sin(vv)),n2)


n1=2
n2=0.5
X3=(R+r*np.sign(np.cos(vv))*np.power(abs(np.cos(vv)),n2))*np.sign(np.cos(uu))*np.power(abs(np.cos(uu)),n1)
Y3=(R+r*np.sign(np.cos(vv))*np.power(abs(np.cos(vv)),n2))*np.sign(np.sin(uu))*np.power(abs(np.sin(uu)),n1)
Z3=r*np.sign(np.sin(vv))*np.power(abs(np.sin(vv)),n2)



fig=plt.figure(figsize=(16,16))

ax = fig.add_subplot(2, 2, 1, projection='3d')
ax.plot_surface(X, Y, Z, cmap=cm.jet, linewidth=0, antialiased=False)
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_zlim(-1, 1)
plt.title('Torus')

ax = fig.add_subplot(2, 2, 2, projection='3d')
ax.plot_surface(X1, Y1, Z1, cmap=cm.jet, linewidth=0, antialiased=False)
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_zlim(-1, 1)
plt.title('Super-toroid $n_{1}=3,  n_{2}=3$')

ax = fig.add_subplot(2, 2, 3, projection='3d')
ax.plot_surface(X2, Y2, Z2, cmap=cm.jet, linewidth=0, antialiased=False)
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_zlim(-1, 1)
plt.title('Super-toroid $n_{1}=3,  n_{2}=0.5$')

ax = fig.add_subplot(2, 2, 4, projection='3d')
ax.plot_surface(X3, Y3, Z3, cmap=cm.jet, linewidth=0, antialiased=False)
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_zlim(-1, 1)
plt.title('Super-toroid $n_{1}=2,  n_{2}=0.5$')
print('  ')
```

## Klein bottle (parametric system I)


<span style="color:#4169E1"> Like the Möbius strip, the Klein bottle is a two-dimensional manifold which is not orientable. Unlike the Möbius strip, it is a closed manifold, meaning it is a compact manifold without boundary. <br>
While the Möbius strip can be embedded in three-dimensional Euclidean space $R^{3}$, the Klein bottle cannot. It can be embedded in $R^{4}$.
<br> 
The Klein bottle can be immersed in a 3D space BUT **this cannot be done without allowing the surface to intersect itself.** 
We present a 3D immersion.  <br>   
The following parametric system, with 2 domains, can be found on Paul Bourque's website.      http://paulbourke.net/geometry/toroidal/

```python
#from mpl_toolkits.mplot3d import axes3d
#from matplotlib import pyplot

resol=199
#    0 <= u < π
#    0 <= v < 2π
u,v =np.meshgrid(np.linspace(start=0,stop=np.pi,num=resol), np.linspace(start=0, stop= 2*np.pi, num=resol))  
r1 = 4*(1 - np.cos(u)/2)   

x1 = 6*np.cos(u)*(1 + np.sin(u)) + r1*np.cos(u)*np.cos(v)
y1 = 16*np.sin(u) + r1*np.sin(u)*np.cos(v)
z1 = r1*np.sin(v)

#    π <= u < 2π
#    0 <= v < 2π

u,v =np.meshgrid(np.linspace(start=np.pi,stop=2*np.pi,num=resol), np.linspace(start=0, stop= 2*np.pi, num=resol)) 
r2 = 4*(1 - np.cos(u)/2)  
x2= 6*np.cos(u)*(1 + np.sin(u)) + r2*np.cos(v + np.pi)
y2= 16*np.sin(u)
z2 = r2*np.sin(v)

rs=4
cs=rs

fig=plt.figure(figsize=(16,8))

ax = fig.add_subplot(1, 2, 1, projection='3d')
surf = ax.plot_surface(x1, y1, z1,  cmap = cm.jet,  antialiased=False, shade=True)
#surf.set_facecolor((0,0,0,0))
surf = ax.plot_surface(x2, y2, z2, cmap = cm.jet,  antialiased=False, shade=True)
#surf.set_facecolor((0,0,0,0))


ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('Klein bottle (3D surface)')

ax = fig.add_subplot(1, 2, 2, projection='3d')
wf=ax.plot_wireframe(x1,y1,z1, rstride=rs, cstride=cs, linewidth=0.15, color='blue')
wf=ax.plot_wireframe(x2,y2,z2, rstride=rs, cstride=cs, linewidth=0.15, color='blue')                       
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.title('Klein bottle (wireframe)')
plt.show()
```

```python
X= np.concatenate((x1, x2), axis=1)
Y= np.concatenate((y1, y2), axis=1)
Z= np.concatenate((z1, z2), axis=1)
```

## Klein bottle (parametric system II)

<span style="color:#4169E1"> From Wikipedia "Kelin bottle" page: In this case the 3D surface is defined with a single parametric system on domain (u,v) with: 
0 ≤ u < π and 0 ≤ v < 2π.

\begin{aligned}x(u,v)=-&{\frac {2}{15}}\cos u\left(3\cos {v}-30\sin {u}+90\cos ^{4}{u}\sin {u}\right.-\\&\left.60\cos ^{6}{u}\sin {u}+5\cos {u}\cos {v}\sin {u}\right)\\y(u,v)=-&{\frac {1}{15}}\sin u\left(3\cos {v}-3\cos ^{2}{u}\cos {v}-48\cos ^{4}{u}\cos {v}+48\cos ^{6}{u}\cos {v}\right.-\\&60\sin {u}+5\cos {u}\cos {v}\sin {u}-5\cos ^{3}{u}\cos {v}\sin {u}-\\&\left.80\cos ^{5}{u}\cos {v}\sin {u}+80\cos ^{7}{u}\cos {v}\sin {u}\right)\\z(u,v)=&{\frac {2}{15}}\left(3+5\cos {u}\sin {u}\right)\sin {v}\end{aligned} <br>
